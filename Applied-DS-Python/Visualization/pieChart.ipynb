# PIE CHART - DEMO

```python
# Import required libraries
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Job data in percentile
job_data = ['40','20','17','8','5','10']

# Define lable as different departments
lables = 'IT', 'Finance','Marketing','Admin','HR','Operations'

# Explode the 1st slice which is IT
explode = (.05,0,0,0,0,0)

# Draw the Pie chart and set the parameters
plt.pie(job_data, labels= lables , explode= explode, )

#Show the plot
plt.show()
```
